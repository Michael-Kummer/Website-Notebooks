```python
clean_data = volunteerOnline22.dropna(subset=[
    "[Bucketed] In the past 12 months, on average, about how many hours per month did you spend formally volunteering? [Do formal volunteering]"
])
clean_data = clean_data.rename(columns={"[Bucketed] In the past 12 months, on average, about how many hours per month did you spend formally volunteering? [Do formal volunteering]":
                                "Hours Spent Per Month Formally Volunteering"})

clean_data.loc[:,"Hours Spent Per Month Formally Volunteering"] = clean_data.loc[:,"Hours Spent Per Month Formally Volunteering"].replace({
    "(0) 0": "0",
    "(1) 1": "1",
    "(2) 2": "2", 
    "(3) 3-5": "3-5",
    "(4) 6-10": "6-10",
    "(5) 11-20": "11-20",
    "(6) 21-40": "21-40",
    "(7) 40+": "40+"
})

hours_order = ["0", "1", "2", "3-5", "6-10", "11-20", "21-40", "40+"]

clean_data["Hours Spent Per Month Formally Volunteering"] = pd.Categorical(
    clean_data["Hours Spent Per Month Formally Volunteering"],
    categories=hours_order,
    ordered=True
)

# (
#     ggplot(clean_data)
#     + geom_bar(aes(
#         x="Hours Spent Per Month Formally Volunteering",
#         fill="What age range do you fall into?"
#       ),
#       position="dodge" 
#     )
#     + coord_flip()
# )
```

# Volunteering Behaviour Analysis

```python
fig, ax = plt.subplots(figsize=(12, 8))  # Increase figure size

# Define the correct order for hours categories using your exact strings
hours_order = [
    "0", "1", "2", "3-5", "6-10", "11-20", "21-40", "40+"
]

# Define the correct order for age ranges using your exact strings
age_order = [
    "18 to 24", "25 to 34", "35 to 44", "45 to 54", "55 to 64", "65 or older"
]

# Filter to only include categories that exist in your data
hours_categories = [h for h in hours_order if h in clean_data["Hours Spent Per Month Formally Volunteering"].unique()]
age_ranges = [a for a in age_order if a in clean_data["What age range do you fall into?"].unique()]

# Group the data
data_grouped = clean_data.groupby(["Hours Spent Per Month Formally Volunteering",
                                 "What age range do you fall into?"]).size().unstack()

# Reindex to ensure correct order
data_grouped = data_grouped.reindex(hours_categories)
data_grouped = data_grouped.reindex(columns=age_ranges)

# Increase bar spacing and sizing
bar_width = 0.95 / len(age_ranges)  # Increased from 0.8
positions = np.arange(len(hours_categories))
positions = positions * 1.5  # Add more vertical spacing between categories

# Add custom colors for better visual distinction
colors = plt.cm.viridis(np.linspace(0, 1, len(age_ranges)))

# Plot each age range separately
for i, (age, color) in enumerate(zip(age_ranges, colors)):
    # Offset each age group's bars
    y_positions = positions + (i - len(age_ranges)/2 + 0.5) * bar_width
    values = data_grouped[age]
    ax.barh(y=y_positions, width=values, height=bar_width, label=age, color=color, edgecolor='black', linewidth=0.5)

ax.set_yticks(positions)
ax.set_yticklabels(hours_categories, fontsize=11)
ax.set_xlabel('Count', fontsize=12)
ax.set_ylabel('Volunteer Hours Per Month', fontsize=12)
ax.set_title('Histogram of Albertans Volunteer hours per month', fontsize = 24)
ax.legend(title='Age Range', fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(axis='x', linestyle='--', alpha=0.7)  

ax.set_ylim(min(positions) - 1, max(positions) + 1)

plt.tight_layout()
plt.show()
```
